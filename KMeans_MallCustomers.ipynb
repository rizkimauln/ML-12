# Modul 11: K-Means Clustering (Adaptasi Dataset Mall Customers)

Dataset yang digunakan adalah **Mall_Customers.csv**.
Langkah pertama adalah melakukan import library yang dibutuhkan.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

## Pengumpulan dan Persiapan Data

Melakukan load data sehingga data tersebut diubah formatnya ke dalam data frame.

```python
baca = pd.read_csv("Mall_Customers.csv")
baca.head()
```

Melihat informasi tentang dataset.

```python
baca.info()
```

Kolom `CustomerID` dan `Gender` dikurangi (di-drop) karena tidak dipergunakan dalam penentuan pada clustering (dan untuk menghindari error akibat data kategori saat scaling).

```python
baca = baca.drop(["CustomerID", "Gender"], axis=1)
baca.head()
```

Menentukan variabel yang akan diklusterkan.
*(Berdasarkan format modul, kita menggunakan iloc[:, 1:3] yang akan mengambil kolom `Annual Income (k$)` dan `Spending Score (1-100)`)*

```python
# Menentukan variabel yang akan di klusterkan
baca_x = baca.iloc[:, 1:3]
baca_x.head()
```

Visualisasi persebaran data awal.

```python
# Memvisualkan persebaran data
plt.scatter(baca['Annual Income (k$)'], baca['Spending Score (1-100)'], s=10, c="red", marker="o", alpha=0.5)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Sebaran Data Awal")
plt.show()
```

Mengubah variabel data menjadi array untuk proses klustering.

```python
x_array = np.array(baca_x)
print(x_array)
```

Melakukan normalisasi data menggunakan `MinMaxScaler`.

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

## Membuat Model Klasifikasi / Clustering

Membuat model KMeans dengan 3 kluster (Sama seperti parameter modul asli).

```python
# Membuat model KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

# Training model 
kmeans.fit(x_scaled)

# Menampilkan label cluster
print(kmeans.labels_)

# Menambahkan hasil cluster ke dataframe
baca["kluster"] = kmeans.labels_

# Menampilkan data
baca.head()
```

## Membuat Report Hasil Klasifikasi

*Catatan: Modul asli menampilkan kode untuk metrik klasifikasi (Accuracy, Precision, Recall, F1-Score). Namun, karena K-Means adalah metode Unsupervised Learning dan kita tidak memiliki label asli, metrik klasifikasi tidak bisa digunakan secara langsung tanpa adanya label kebenaran. Sebagai gantinya, kita memvisualisasikan hasil kluster.*

Visualisasi dari kluster beserta dengan plot data centroid.

```python
# Plot hasil clustering
plt.figure(figsize=(8,6))

plt.scatter(
    baca['Annual Income (k$)'],
    baca['Spending Score (1-100)'],
    c=baca['kluster'],
    cmap='viridis',
    alpha=0.6
)

# Titik centroid (perlu dikembalikan ke skala awal agar sesuai dengan plot)
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

plt.scatter(
    centroids[:, 0],   # Annual Income
    centroids[:, 1],   # Spending Score
    s=200,
    c='red',
    marker='X',
    label='Centroid'
)

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering - Mall Customers')
plt.legend()

plt.show()
```
